**EDA**

```python
data.groupby('Target')['Target'].count()
```

```python
pd.crosstab(data.Target,['Target']).plot(kind='pie', subplots=True,autopct='%1.1f%%')
```

**Demographic Factors**

```python
#Gender distribution
pd.crosstab(data.Target,data.Gender)
```

```python
pd.crosstab(data.Target,data.Gender).plot(kind='bar',title ='Gender distribution')
```

From the data, we can see that:
- There are more female students than male students
- There are 2x more female graduates than male graduates

```python
#Age 
#VIew the age students mostly enroll for higher education
plt.hist(data['Age at enrollment'])
plt.xlabel('Age')
plt.ylabel('Count of students')
plt.title('Distribution of enrollment age')
#How does this contribute to their education?
sns.catplot(x='Target',y='Age at enrollment',kind = 'violin',data =data).set(title ='Education Outcome based on age at enrollment')
```

- Most students enroll between the ages of 18 - 20. the rate of enrollment declines steadily right after.
- There are more graduates among students who enrolled in the early years of 17 - 23. The rate of droputs increase from students who enrolled at 24 years upwards.

**Socio-Economic Factors**

```python
#Mother's Occupation
pd.crosstab(data["Mother's occupation"],data['Target']).plot(kind='bar',title = ("Effect of Mother's occupation"))
```

- There's a high rate of graduating among Students who have mothers in the unskilled workers category(9), and Adiministrative staff category(4), while students with mothers who are also students(0), or in the other situation category(90) have a higher rate of dropping out of school.

```python
#Status of Tuition
pd.crosstab(data['Target'],data['Tuition fees up to date'])
```

```python
pd.crosstab(data['Target'],data['Tuition fees up to date']).plot(kind='bar',title='Effect of Tuition availabilty')
```

- Students whose tuition are up to date have a higher rate of graduating than students with incomplete tuition.

```python
#Course of study
pd.crosstab(data["Course"],data['Target'])
```

```python
pd.crosstab(data['Course'],data['Target']).plot(kind='bar', title = 'Choice of course of study')
```

- Nursing students(9500) have the highest rate of graduating, followed by Social services students(9238). Management students(9991) with evening attendance, have higher dropout rates, with more dropouts than graduates.

```python
#Unemployment rate
pd.crosstab(data['Unemployment rate'],data['Target'])
```

```python
sns.catplot(x='Target',y='Unemployment rate',kind = 'box',data =data).set(title ='Effect of Unemployment rate')
```

- Countries with lower unemployment rates have more graduates than Countries with higher unemployment rates.
